# Stochastic Processes

Anda adalah seorang analis keuangan yang sedang mempelajari pergerakan harga saham XYZ Inc. selama 10 hari terakhir. Berikut adalah data historis harga saham XYZ Inc. selama 10 hari terakhir:

[100,105,110,115,112,108,115,118,120,125]

Anda ingin menggunakan model stokastik untuk memprediksi harga saham XYZ Inc. selama 5 hari ke depan. Gunakan model stokastik dengan parameter sebagai berikut:
image.png

Lakukan prediksi harga saham XYZ Inc. untuk 5 hari ke depan menggunakan model stokastik.
Gambarkan grafik pergerakan harga saham XYZ Inc. selama 15 hari terakhir, termasuk data historis dan prediksi yang Anda buat.

```python
# Data historis harga saham selama 10 hari
historical_prices = np.array([100, 105, 110, 115, 112, 108, 115, 118, 120, 125])

# Parameter model stokastik
growth = 0.1    # Rata-rata pertumbuhan
sigma = 0.2     # Volatilitas
dt = 1          # Interval waktu (1 hari)
T = 15          # Waktu akhir (15 hari)
n_steps = T - len(historical_prices)  # Jumlah langkah prediksi (5 hari)

# Inisialisasi array harga saham dengan data historis
prices = list(historical_prices)

# Gunakan metode Random Number Simulation untuk mensimulasikan prediksi harga saham
current_price = historical_prices[-1]
for _ in range(n_steps):
    random_shock = np.random.normal(0, 1)  # Menghasilkan angka acak dari distribusi normal standar
    next_price = current_price * np.exp((growth - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * random_shock)
    prices.append(next_price)
    current_price = next_price

# Buat plot pergerakan harga saham selama 15 hari
days = np.arange(1, 16)
plt.figure(figsize=(10, 5))
plt.plot(days[:10], historical_prices, marker='o', label='Harga Historis')
plt.plot(days, prices, marker='x', linestyle='--', color='orange', label='Prediksi Harga')
plt.title('Prediksi Harga Saham XYZ Inc. Menggunakan Model Stokastik')
plt.xlabel('Hari ke-')
plt.ylabel('Harga Saham')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Distribusi Uniform

Sebuah perusahaan logistik bertanggung jawab untuk mengelola pengiriman paket dari pusat distribusi ke berbagai titik tujuan. Untuk meningkatkan efisiensi operasional, perusahaan menggunakan pendekatan distribusi uniform untuk menentukan waktu pengiriman paket kepada pelanggan.

Deskripsi Kasus:
Perusahaan logistik memiliki 3 pusat distribusi yang berlokasi di kota besar A, B, dan C. Setiap pusat distribusi bertanggung jawab untuk mengirimkan paket ke 5 tujuan yang berbeda. Pengiriman dilakukan setiap hari pada jam yang sama untuk memastikan konsistensi dalam layanan.

Detail Kasus:
1. Waktu Pengiriman:

Waktu pengiriman dari setiap pusat distribusi diatur dengan distribusi uniform antara pukul 9 pagi hingga pukul 5 sore.
Setiap titik tujuan memiliki interval waktu yang berbeda untuk pengiriman, yang dipilih secara acak tetapi merata dalam rentang waktu tersebut.

2. Tujuan Pengiriman:

Titik tujuan A1, A2, dan A3 berada di kota besar A.
Titik tujuan B1, B2, dan B3 berada di kota besar B.
Titik tujuan C1, C2, dan C3 berada di kota besar C.

3. Jadwal Pengiriman:

Setiap titik tujuan menerima satu pengiriman setiap hari.
Perusahaan ingin memastikan bahwa waktu pengiriman paket ke setiap titik tujuan adalah acak dalam rentang waktu yang ditentukan, tetapi merata untuk menghindari kemacetan lalu lintas dan mengoptimalkan waktu pengiriman.

Pertanyaan:

1. Tentukan rentang waktu untuk setiap titik tujuan di kota A, B, dan C untuk menerima pengiriman paket. Hitunglah interval waktu yang merata dari 9 pagi hingga 5 sore.

2. Berapa peluang bahwa paket akan tiba di titik tujuan A1 antara pukul 10 pagi hingga 12 siang? Hitung peluangnya menggunakan distribusi uniform.

3. Apa peluang bahwa paket akan tiba di titik tujuan B3 antara pukul 2 hingga 4 sore? Hitung peluangnya menggunakan distribusi uniform.

```python
# Waktu pengiriman dalam menit (9 pagi = 540, 5 sore = 1020)
start_time = 540  # 9:00 dalam menit
end_time = 1020  # 17:00 dalam menit
total_minutes = end_time - start_time

# Fungsi untuk menghitung interval waktu yang merata (dalam menit)
def delivery_interval(start, end):
    return np.random.uniform(start, end)

# Rentang waktu untuk setiap titik tujuan
destinations = {
    'A1': delivery_interval(start_time, end_time),
    'A2': delivery_interval(start_time, end_time),
    'A3': delivery_interval(start_time, end_time),
    'B1': delivery_interval(start_time, end_time),
    'B2': delivery_interval(start_time, end_time),
    'B3': delivery_interval(start_time, end_time),
    'C1': delivery_interval(start_time, end_time),
    'C2': delivery_interval(start_time, end_time),
    'C3': delivery_interval(start_time, end_time),
}

# 1. Cetak rentang waktu pengiriman untuk setiap titik tujuan (dalam jam dan menit)
print("Rentang Waktu Pengiriman (jam:menit):")
for dest, time in destinations.items():
    hours = int(time // 60)  # Menghitung jam
    minutes = int(time % 60)  # Menghitung menit
    print(f"{dest}: {hours:02d}:{minutes:02d}")

# 2. Hitung peluang paket tiba di A1 antara pukul 10:00 dan 12:00
prob_a1 = (720 - 600) / total_minutes  # 10:00 = 600, 12:00 = 720
print(f"\nPeluang paket tiba di A1 antara 10:00-12:00: {prob_a1}")

# 3. Hitung peluang paket tiba di B3 antara pukul 14:00 dan 16:00
prob_b3 = (960 - 840) / total_minutes  # 14:00 = 840, 16:00 = 960
print(f"Peluang paket tiba di B3 antara 14:00-16:00: {prob_b3}")
```

# LCG

Sebuah perusahaan pengembangan permainan sedang merencanakan untuk menggunakan Linear Congruential Generator (LCG) untuk menghasilkan angka acak dalam pengembangan permainan mereka.

Deskripsi Kasus:
Perusahaan permainan ini ingin menggunakan LCG untuk menghasilkan urutan angka acak yang akan digunakan dalam berbagai konteks dalam permainan mereka, seperti posisi pemain, kekuatan senjata, atau hadiah.

Detail Kasus:

Parameter LCG:
    Perusahaan telah memilih parameter LCG sebagai berikut:
        image.png
        
Penggunaan LCG:
Perusahaan akan menggunakan LCG untuk menghasilkan urutan angka acak dalam rentang 0 hingga p-1

Setiap kali permainan memerlukan angka acak baru, LCG akan digunakan dengan angka yang dihasilkan sebagai masukan untuk menghasilkan angka acak selanjutnya.

Pertanyaan:

image.png


```python
# Parameter LCG
p = 2**31 - 1  # Modulus
a = 16807      # Multiplier
c = 0          # Increment

# Fungsi untuk menghasilkan angka acak menggunakan LCG
def lcg(x0, n):
  xn = x0
  results = []
  for _ in range(n):
    xn = (a * xn + c) % p
    results.append(xn)
  return results

# 1. Menghitung nilai X10 jika X0 = 7
x0 = 7
x10_results = lcg(x0, 10)  # Menghasilkan urutan X0 hingga X9
x10 = x10_results[-1] # Mengambil hasil perhitungan LCG untuk X10

print(f"Nilai X10 jika X0 = 7 adalah: {x10}")

# 2. Apa yang terjadi jika nilai X0 diubah menjadi 0?
x0 = 0
x_results_0 = lcg(x0,10)
x10_0 = x_results_0[-1]

print(f"Nilai X10 jika X0 = 0 adalah: {x10_0}")

#Penjelasan:
'''
Jika nilai X0 adalah 0 maka hasil dari X1 akan selalu 0 karena 0 dikalikan dengan berapapun hasilnya akan selalu 0.
Nilai c = 0 sehingga tidak akan mempengaruhi hasil perhitungan.
Sehingga semua nilai Xn selanjutnya akan bernilai 0
'''
```
