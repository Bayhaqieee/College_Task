# Monte Carlo

## Soal 1

### Simulasi Monte Carlo: Prediksi Permintaan Sepatu

Sebuah toko sepatu memperkirakan permintaan sepatu per harinya menurut pola distribusi sebagai berikut:

| No. Urut | Permintaan per Hari | Frekuensi Permintaan |
|----------|----------------------|-----------------------|
| 1        | 4 pasang             | 5                     |
| 2        | 5 pasang             | 10                    |
| 3        | 6 pasang             | 15                    |
| 4        | 7 pasang             | 30                    |
| 5        | 8 pasang             | 25                    |
| 6        | 9 pasang             | 15                    |
|          | **Jumlah**           | **100**               |

---

### Instruksi

Perusahaan ingin memprediksi **permintaan selama 5 hari ke depan**.  
Gunakan **Simulasi Monte Carlo** untuk memperkirakan jumlah permintaan sepatu berdasarkan data distribusi di atas.

---

```python
# Menentukan distribusi probabilitas
permintaan = [4, 5, 6, 7, 8, 9]
frekuensi = [5, 10, 15, 30, 25, 15]
probabilitas = [f/sum(frekuensi) for f in frekuensi]

# Membuat distribusi probabilitas kumulatif
probabilitas_kumulatif = np.cumsum(probabilitas)

# Menentukan interval bilangan acak
interval_acak = [[0, probabilitas_kumulatif[0]]]
for i in range(1, len(probabilitas_kumulatif)):
    interval_acak.append([probabilitas_kumulatif[i-1], probabilitas_kumulatif[i]])

# Fungsi untuk simulasi satu hari (dengan menampilkan angka random)
def simulasi_hari():
    bilangan_acak = np.random.randint(0, 30)
    print(f"Bilangan acak: {bilangan_acak}")  # tampilkan angka random

    for i in range(len(interval_acak)):
        if interval_acak[i][0] * 100 <= bilangan_acak < interval_acak[i][1] * 100:
            return permintaan[i]
    return permintaan[0]

# Melakukan simulasi selama 5 hari
hasil_simulasi = []
for _ in range(5):
    hasil_simulasi.append(simulasi_hari())

print("----------------------------------------------------------------------------")

# Visualisasi
plt.figure(figsize=(8, 6))
plt.bar(range(1, 6), hasil_simulasi, color='skyblue')
plt.xlabel('Hari')
plt.ylabel('Permintaan Sepatu (pasang)')
plt.title('Simulasi Permintaan Sepatu Selama 5 Hari')
plt.xticks(range(1, 6))
for i, v in enumerate(hasil_simulasi):
  plt.text(i + 1 - 0.2, v + 0.2, str(v), color='black')
plt.show()

# Total Permintaan
total_permintaan = sum(hasil_simulasi)
print("----------------------------------------------------------------------------")
print(f"Total Permintaan Sepatu selama 5 hari: {total_permintaan} pasang")
```
